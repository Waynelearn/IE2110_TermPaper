```python
from queue import PriorityQueue
import networkx as nx
import matplotlib.pyplot as plt
import random as r

import folium
import numpy as np
import pandas as pd
from pulp import *
import pandas as pd
```

```python
class node(object):
    def __init__(self,code,name,pos):
        self.code=code
        self.name=name
        self.dist={}
        self.pos=pos

    def add_edge(self,new_node,weight):
        self.dist[new_node]=weight
```

```python
class Graph:

    def __init__(self, node_lst):
        self.v=len(node_lst)
        self.node_lst = node_lst
        self.edges = [[-1 for i in range(len(node_lst))] for j in range(len(node_lst))]
        self.visited = []
        
    def add_all_edges(self):
        for c in self.node_lst:
            for d in c.dist:
                self.edges[c.code][d.code]=c.dist[d]
                self.edges[d.code][c.code]=c.dist[d]

    
    def dijkstra(self, start_vertex):
        D = {v:float('inf') for v in range(self.v)}
        D[start_vertex] = 0

        pq = PriorityQueue()
        pq.put((0, start_vertex))

        while not pq.empty():
            (dist, current_vertex) = pq.get()
            self.visited.append(current_vertex)

            for neighbor in range(self.v):
                if self.edges[current_vertex][neighbor] != -1:
                    distance = self.edges[current_vertex][neighbor]
                    if neighbor not in self.visited:
                        old_cost = D[neighbor]
                        new_cost = D[current_vertex] + distance
                        if new_cost < old_cost:
                            pq.put((new_cost, neighbor))
                            D[neighbor] = new_cost
        return D
    
    def plot(self):
        G=nx.Graph()
        for c in self.node_lst:
            G.add_node(c.name,pos=c.pos)
        for c in self.node_lst:
            for d in c.dist:
                G.add_edge(c.name,d.name,weight=c.dist[d])
        pos=nx.get_node_attributes(G,'pos')

        nx.draw(G, pos,with_labels=True,node_size=1000,node_color='r')
        labels = nx.get_edge_attributes(G,'weight')
        nx.draw_networkx_edge_labels(G,pos,edge_labels=labels)
        plt.show()
    
    def transport_cost_matrix(self):
        names=[c.name for c in self.node_lst]
        transport=pd.DataFrame(np.zeros((self.v,self.v)),index=names,columns=names)
        for i in self.node_lst:
            g=Graph(self.node_lst)
            g.add_all_edges()
            D=g.dijkstra(i.code)
            for j in D:
                transport.loc[i.name,self.node_lst[j].name]=D[j]
        return transport
```

# Create initial Network Model

```python
#Node to node transport cost
df=pd.read_csv('routes_c.csv',index_col=0)
df
```

# Read in raw csv data

```python
#production cost matrix
prod_cost=pd.read_csv('prod_cost.csv',index_col=0)
prod_cost
```

```python
#fixed cost matrix
fixed_cost=pd.read_csv('fixed_cost.csv',index_col=0)
fixed_cost
```

```python
#capacity matrix
cap=pd.DataFrame({'Low_Cap':[500000 for i in range(len(loc))],
                 'High_Cap':[1500000 for i in range(len(loc))]},index=loc)
cap
```

```python
#demand matrix
demand=pd.read_csv('demand.csv',index_col=0)
demand
```

# Initial Base Model LP

```python
# Initialize, Define Decision Vars., and Objective Function
model = LpProblem("Capacitated_Plant_Location_Model", LpMinimize)
x = LpVariable.dicts("production_", [(i,j,s) for i in loc for j in loc for s in size],
                     lowBound=0, upBound=None, cat='Continuous')
y = LpVariable.dicts("plant_", 
                     [(i,s) for s in size for i in loc], cat='Binary')
model += (lpSum([fixed_cost.loc[i,s] * y[(i,s)] for s in size for i in loc])
          + lpSum([(prod_cost.loc[i,s]+transport.loc[i,j]) * x[(i,j,s)] for i in loc for j in loc for s in size]))

# Define the constraints
for j in loc:
    model += lpSum([x[i,j,s] for i in loc for s in size]) == demand.loc[j,'Dmd']
    
for i in loc:
    for s in size:
        model += lpSum([x[(i, j, s)] for j in loc]) <= lpSum([cap.loc[i,s]*y[(i,s)] ])
model.writeLP('initial_base_case.txt')
```

```python
#solve the LP
model.solve()
print("Status:", LpStatus[model.status])
print(f'objective:{value(model.objective)}')
```

### Solution for $x_{ijs}$

```python
#Result table for x_ijs
result = np.array([[value(x[i,j,s]) for i in loc for s in size] for j in loc])
t_loc=[]
for i in loc:
    t_loc.append(i)
    t_loc.append(i)
result_table=pd.DataFrame(result,index=loc,columns=pd.MultiIndex.from_tuples(zip(t_loc,size*len(loc))))
result_table
```

### Solution for $y_{is}$

```python
plant=[]
for i in loc:
    for s in size:
        plant.append(value(y[i,s]))
plant_open=pd.DataFrame(plant,index=pd.MultiIndex.from_tuples(zip(t_loc,size*len(loc))),columns=['Open'])
plant_open
```

### Shadow price for base model

```python
#Shadow price of initial base model
o=[{'name':name,'shadow price':c.pi,'slack':{c.slack}} for name,c in model.constraints.items()]
shadow=pd.DataFrame(o)
```

### Average unit cost of production (includes fixed cost)

```python
avg_cost=[]
for i in loc:
    for s in size:
        plant=value(y[i,s])*fixed_cost.loc[i,s]
        total=plant
        products=0
        for j in loc:
            total+=value(x[i,j,s])*prod_cost.loc[i,s]
            products+=value(x[i,j,s])
        if products!=0:
            avg_cost.append(total/products)
        else:
            avg_cost.append(0)
avg_cost
cost=pd.DataFrame(avg_cost,index=pd.MultiIndex.from_tuples(zip(t_loc,size*len(loc))),columns=['Average Cost'])
cost
```

### Cost breakdown for supplying region j

```python
total_cost_lst=[]
transport_cost_lst=[]
prod_cost_lst=[]
for j in loc:
    total_cost=0
    #products=0
    transport_cost=0
    production_cost=0
    for i in loc:
        for s in size:
            products+=value(x[i,j,s])
            total_cost+=value(x[i,j,s])*cost.loc[i].loc[s]['Average Cost']+value(x[i,j,s])*transport.loc[i,j]
            transport_cost+=value(x[i,j,s])*transport.loc[i,j]
            production_cost+=value(x[i,j,s])*cost.loc[i].loc[s]['Average Cost']
    total_cost_lst.append(total_cost)
    transport_cost_lst.append(transport_cost)
    prod_cost_lst.append(production_cost)
    
master=zip(transport_cost_lst,prod_cost_lst,total_cost_lst)
total_cost_base=pd.DataFrame(master,index=loc,columns=['transport_cost','production_cost','Total_cost'])
total_cost_base
```

# Case 3: Disrupted operations within certain regions due to natural disasters, strikes, pandemics, and other environmental/socio-political reasons.

```python
obj_lst=[]
for country in loc:
    # Initialize, Define Decision Vars., and Objective Function
    model4 = LpProblem("Capacitated_Plant_Location_Model", LpMinimize)
    x4 = LpVariable.dicts("production_", [(i,j,s) for i in loc for j in loc for s in size],
                         lowBound=0, upBound=None, cat='Continuous')
    y4 = LpVariable.dicts("plant_", 
                         [(i,s) for s in size for i in loc], cat='Binary')
    model4 += (lpSum([fixed_cost.loc[i,s] * y4[(i,s)] for s in size for i in loc])
              + lpSum([(prod_cost.loc[i,s]+transport.loc[i,j]) * x4[(i,j,s)] for i in loc for j in loc for s in size]))

    # Define the constraints
    for j in loc:
        model4 += lpSum([x4[i,j,s] for i in loc for s in size]) == demand.loc[j,'Dmd']

    for i in loc:
        for s in size:
            model4 += lpSum([x4[(i, j, s)] for j in loc]) <= lpSum([cap.loc[i,s]*y4[(i,s)] ])
    
    ####################set some factories to closed############################
    for s in size:
        model4 += y4[country,s] == 0
    ############################################################################
    model4.solve()
    obj_lst.append(value(model4.objective)-value(model.objective))

df3=pd.DataFrame(obj_lst,index=loc,columns=['Difference in Objective'])
df3
```
